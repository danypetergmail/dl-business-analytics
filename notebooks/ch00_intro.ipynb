## About This Book

Deep learning is no longer the exclusive domain of computer scientists and machine learning researchers. It has become a practical tool that business analysts, operations managers, and strategy consultants encounter every day — embedded in demand forecasting systems, customer churn models, fraud detection pipelines, and the large language model assistants that are reshaping how knowledge work gets done.

This book was written for **students and professionals who work with data** but did not necessarily train as software engineers. The assumption is that you are comfortable reasoning about a dataset, have written some Python, and have some exposure to NumPy or pandas — but that words like *backpropagation*, *convolutional kernel*, or *attention mechanism* are still unfamiliar territory. The goal is to make them familiar, and to do so through practice rather than theory alone.

Every chapter in this book is a Jupyter notebook (`.ipynb`). Explanations, code, and exercises all live together so that reading and doing are never separated. All datasets used are publicly available and free to download. The notebooks are designed to run in **Google Colab**, but they work equally well in any standard Jupyter environment — on your own machine, your university's computing cluster, or any cloud platform that supports Python notebooks.

## Getting Your Environment Ready

The cells below install all required packages and verify your environment. **Run them once before starting Chapter 1.** They take about 5 minutes on a fresh environment.

If you are on **Google Colab**: the installation cell handles everything automatically.  
If you are on a **local machine**: ensure you have Python 3.9+ and pip available, then run the same cell.

### Step 1 — Install packages

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 1 — Install all required packages
# Run this cell once. It installs everything needed across all chapters.
# Expected time: 2–5 minutes on a fresh environment.
# ─────────────────────────────────────────────────────────────────────────────

import subprocess, sys

packages = [
    "torch",           # Deep learning framework (Chapters 2–6)
    "torchvision",     # Image datasets and transforms (Chapters 4–5)
    "torchtext",       # Text utilities (Chapter 6)
    "numpy",           # Numerical computing
    "pandas",          # Tabular data handling
    "matplotlib",      # Plotting and visualisation
    "seaborn",         # Statistical plots
    "scikit-learn",    # Preprocessing, metrics, classical ML baselines
    "transformers",    # Hugging Face — tokenizers and models (Chapters 6–7)
    "datasets",        # Hugging Face — dataset loading (Chapters 6–7)
    "gradio",          # Simple web interfaces for demos (Chapter 7)
    "kaggle",          # Kaggle API for dataset downloads (Chapters 2–6)
    "ipywidgets",      # Interactive widgets in notebooks
    "tqdm",            # Progress bars during training
]

print("Installing packages...\n")
for pkg in packages:
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", "--quiet", pkg],
        capture_output=True, text=True
    )
    status = "OK" if result.returncode == 0 else "FAILED"
    print(f"  {pkg:<20} {status}")

print("\nInstallation complete. Proceed to Step 2.")
````

### Step 2 — Verify your environment

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 2 — Verify environment
# Checks Python, PyTorch, and key library versions.
# Confirms whether a GPU is available (not required, but noted).
# ─────────────────────────────────────────────────────────────────────────────

import sys
import importlib

print("=" * 55)
print("  Environment Verification")
print("=" * 55)

# Python version
py_version = sys.version.split(" ")[0]
py_status  = "OK" if int(py_version.split(".")[1]) >= 9 else "WARNING: Python 3.9+ recommended"
print(f"  Python           {py_version:<12}  {py_status}")

# Core libraries
libs = [
    ("torch",        "2.0"),
    ("torchvision",  "0.15"),
    ("numpy",        "1.23"),
    ("pandas",       "1.5"),
    ("matplotlib",   "3.6"),
    ("sklearn",      "1.2"),
    ("transformers", "4.30"),
    ("gradio",       "3.0"),
]

for lib_name, min_ver in libs:
    try:
        mod     = importlib.import_module(lib_name)
        version = getattr(mod, "__version__", "unknown")
        print(f"  {lib_name:<16} {version:<12}  OK")
    except ImportError:
        print(f"  {lib_name:<16} {'NOT FOUND':<12}  Re-run Step 1")

# GPU / CPU check
print()
print("-" * 55)
try:
    import torch
    if torch.cuda.is_available():
        gpu_name = torch.cuda.get_device_name(0)
        print(f"  Compute          GPU available: {gpu_name}")
        print(f"  CUDA version     {torch.version.cuda}")
    elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
        print("  Compute          Apple MPS (Metal) available")
    else:
        print("  Compute          CPU only")
        print("  Note: GPU is not required — all exercises run on CPU.")
except Exception as e:
    print(f"  Compute check failed: {e}")

print("=" * 55)
print("  Verification complete.")
print("=" * 55)
````

### Step 3 — Quick PyTorch smoke test

A minimal tensor operation to confirm PyTorch is working correctly on your available device.

````python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 3 — PyTorch smoke test
# Creates two small tensors, multiplies them, and prints the result.
# If this cell runs without error, PyTorch is working correctly.
# ─────────────────────────────────────────────────────────────────────────────

import torch

# Determine the best available device
if torch.cuda.is_available():
    device = torch.device("cuda")
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")

print(f"Using device: {device}")
print()

# Create two simple tensors and perform a matrix multiplication
# We will explain exactly what this does in Chapter 2
# For now, we just want to confirm the operation completes successfully
A = torch.tensor([[1.0, 2.0],
                   [3.0, 4.0]], device=device)

B = torch.tensor([[5.0, 6.0],
                   [7.0, 8.0]], device=device)

C = torch.matmul(A, B)  # Matrix multiplication — covered in depth in Chapter 2

print("Tensor A:")
print(A)
print()
print("Tensor B:")
print(B)
print()
print("A × B =")
print(C)
print()
print("PyTorch smoke test passed. You are ready to begin.")
````
